```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix
from imblearn.over_sampling import SMOTE
```

```python
# read the application data from a csv file
app_data = pd.read_csv("src/application_data.csv",sep=",", encoding='cp1252', index_col=0)
app_data.head()
```

```python
info = {
    "CODE_GENDER": {
        "title": "Gender of the client",
        "xLabel":"Gender",
        "yLabel": "Frequency",
        "figsize": (12,8),
        "rotation": 0
    },
    "NAME_FAMILY_STATUS": {
        "title": "Family status of the client",
        "xLabel": "Family status",
        "yLable": "Frequency",
        "figsize": (12,8),
        "rotation": 25
    },
    "NAME_INCOME_TYPE": {
        "title": "Clients income type",
        "xLabel": "income type",
        "yLable": "Frequency",
        "figsize": (12,8),
        "rotation": 25
    },
    "NAME_EDUCATION_TYPE": {
        "title": "Level of highest education the client achieved",
        "xLabel": "highest education",
        "yLable": "Frequency",
        "figsize": (12,8),
        "rotation": 25
    }
}
```

```python
for cols in info:
    print(cols)
    
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=info[cols]["figsize"])
    plt.subplots_adjust(wspace=0.3, hspace=1)
    
    #Pie plot 1
    data = app_data[cols].value_counts().sort_index()
    ax1.pie(data, labels=data.index, colors=sns.color_palette('flare'), autopct='%.0f%%', startangle=90, textprops={'fontsize': 9})
    ax1.set_title(info[cols]["title"])
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=info[cols]["rotation"])
    
    #Pie plot 2
    target_one = app_data[app_data['TARGET'] == 1][cols]
    data = target_one.value_counts().sort_index()
    ax2.pie(data, labels=data.index, colors=sns.color_palette('Blues'), autopct='%.0f%%', startangle=90)
    ax2.set_title(f"{info[cols]['title']}where target = 1")
    plt.show()
    
    fig, (ax3, ax4) = plt.subplots(1,2, figsize=info[cols]["figsize"])
    plt.subplots_adjust(wspace=0.3, hspace=1)
    
    #Hist plot 1
    sns.histplot(app_data[cols], ax=ax3)
    ax3.set_xticklabels(ax3.get_xticklabels(), rotation=info[cols]["rotation"])
    ax3.set_title(info[cols]["title"])
    ax3.set_xlabel(info[cols]["xLabel"])
    ax3.set_ylabel("Frequency")

    #Hist plot 2
    sns.histplot(app_data, x=cols, hue="TARGET", ax=ax4)
    ax4.set_xticklabels(ax4.get_xticklabels(), rotation=info[cols]["rotation"])
    ax4.set_title(info[cols]["title"])
    ax4.set_xlabel(info[cols]["xLabel"])
    ax4.set_ylabel("Frequency")
    
    plt.show()
    
    # Stacked Bar Chart Percentages 
    percentages = app_data.groupby(cols)['TARGET'].value_counts(normalize=True).unstack().fillna(0)

    # Create a 100% stacked bar chart
    plt.figure(figsize=info[cols]["figsize"])
    percentages.plot(kind='bar', stacked=True, colormap='icefire', ax=plt.gca())
    plt.xlabel(info[cols]["xLabel"])
    plt.ylabel('Percentage')
    plt.title(f'100% Stacked Bar Chart {info[cols]["title"]}')
    plt.legend(title='TARGET', loc='upper left', labels=['0', '1'])
    plt.show()
    
    fig, (ax5, ax6) = plt.subplots(1,2, figsize=info[cols]["figsize"])
    plt.subplots_adjust(wspace=0.3, hspace=1)
    
    #Hist plot Target = 0
    target_zero = app_data[app_data['TARGET'] == 0].sort_values(cols)[cols]
    sns.histplot(target_zero, color="blue", ax=ax5)
    ax5.set_xticklabels(ax5.get_xticklabels(), rotation=info[cols]["rotation"])
    ax5.set_title(f"{info[cols]['title']} where Target = 0")
    ax5.set_xlabel(info[cols]["xLabel"])
    ax5.set_ylabel("Frequency")
    
    
    #Hist plot Target = 1
    target_one = app_data[app_data['TARGET'] == 1].sort_values(cols)[cols]
    sns.histplot(target_one, color="orange", ax=ax6)
    ax6.set_xticklabels(ax6.get_xticklabels(), rotation=info[cols]["rotation"])
    ax6.set_title(f"{info[cols]['title']} where Target = 1")
    ax6.set_xlabel(info[cols]["xLabel"])
    ax6.set_ylabel("Frequency")
    
    plt.show()
```
